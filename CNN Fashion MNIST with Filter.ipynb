```
# convert images from integer to float
train_images = train_images/255.0
test_images = test_images/255.0
```

```
# reshping the images into single channel
train_images = train_images.reshape(len(train_images),28,28,1)
test_images = test_images.reshape(len(test_images),28,28,1)
```

```
# definining the CNN model
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import Dense
from keras.layers import ReLU
```

```
model = Sequential()
```

```
# We start with single convolutional layer with small filter size (3,3) with number of filters = 32
model.add(Conv2D(32, (3,3), activation = "relu", kernel_initializer = "he_uniform", input_shape = (28,28,1)))
```

```
# Add max pooling layer
from keras.layers import MaxPooling2D
```

```
model.add(MaxPooling2D((2, 2)))
```

```
# flattening the filter maps for providing input to the neural classifier
from keras.layers import Flatten
```

```
model.add(Flatten())
```

```
# start neural classifier with 1st layer having 100 neurons
model.add(Dense(100, activation= "relu", kernel_initializer= "he_uniform"))
```

```
# output layer of neural classifier with 10 neurons and softmax activation function
model.add(Dense(10, activation= "softmax"))
```

```
# compile the model with Stochstic Gradient descent model
from keras.optimizers import SGD
opt = SGD(lr = 0.01, momentum = 0.9)
```

```
model.compile(optimizer= opt, loss = "sparse_categorical_crossentropy", metrics= ["accuracy"])
```

```
mnist_model = model.fit(train_images,train_labels,epochs=3,validation_split=0.1)
```
